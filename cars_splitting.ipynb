```python
import pandas as pd
df = pd.read_csv('/content/car_data.csv')
df.dropna()
df.head()
```

```python
df[['Make', 'Model_name']] = df.Model.str.split(" ",1, expand = True)
df.head()
```

```python
del df['Make']
del df['Model']
df.head()
```

```python
df[['Make', 'Model']] = df.Model_name.str.split(" ",1, expand = True)
df.head()
```

```python
del df['Model_name']
df.head()
```

```python
df.insert(2,"Brand", 5)
df.head()
```

```python
df.insert(3,"Model_Name", 5)
df = df.assign(Brand = df.Make)
df = df.assign(Model_Name = df.Model)
df.head()
```

```python
del df['Model']
del df['Make']
```

```python
df.rename(columns = {'Brand':'Make'}, inplace = True)
df.rename(columns = {'Model_Name':'Model'}, inplace = True)
df.head()
```

```python
df.to_csv('/content/cars.csv')
```
